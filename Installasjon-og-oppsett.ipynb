# Installasjon og Oppsett

## Installasjon

Følgende Software installeres
* Python 
* Microsoft Visual Studio Code
* git

Når Python er installert kan man teste at det fungerer ved å åpne en terminal og kjøre følgende kommando
```
py --version
```

### Virtuelt miljø for kurset

Vi bruker et virtuelt environment miljø i kurset. Kjør i terminal. (Powershell anbefales over cmd.exe)
```
py -m venv venv 
# i cmd.exe
venv/Scripts/activate.bat
# i powershell.exe
venv/Scripts/Activate.ps1
```
Du må aktivere venv hver gang du restarter powershell.

Vi trenger også endel 3rd parts biblioteker som vi kan installere med pip. Klon python kurset med git til ønsket mappe
```
git clone git@git.enett.foe:unix/kurs/python.git
Eller eller
git clone https://git.enett.foe/unix/kurs/python.git
```

Kjør følgende kommandoer i terminal fra mappen til python kurset du nettopp klonet. <br>
```
py -m pip install -r requirments.txt
```

Da har du installert alle pakkene du trenger 

## Oppsett

I VSCode anbefaler vi noen extensions, disse kan du finne i extensions fana til venstre. De anbefaler vi er:
* Python
* Jupyter

## Jupyter Notebooks

Kursfilene er skrevet i Jupyter Notebook. For å starte jupyter notebokk skriv følgende i terminalen:    
```
py -m jupyter lab
```
Da starter en nettleser med Jupyter Lab 

## VSCode

Gå til Extensions (Ctrl + Shift + X) Søk etter "Python microsoft" og installer denne. 
Søk så etter "jupyter". Installer "Jupyter" fra Microsoft. 

Da kan vi enkelt kjøre Jupyter Notebooks i VSCode. 


## Kjøre python kode

Skriver du nå "py" i powershell kommer du inn i python shellet. Da kan du starte med
```
print("Hello World!")
```

For å komme ut av python shellet skriv 
```
exit()
```

I VScode kan du opprette en .py fil og kjøre denne med `Ctrl + F5` eller starte en terminal i VScode og skrive "py -m script.py".
